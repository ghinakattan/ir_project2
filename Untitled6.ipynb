```python
import json
import numpy as np
import joblib
from collections import defaultdict
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
import re
base_path = "offline_indexing_service/data/trec_tot"  # ÿ∫Ÿäÿ±Ÿá ÿ≠ÿ≥ÿ® ŸÖŸÉÿßŸÜŸÉ

# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÖŸÑŸÅÿßÿ™
with open(f"{base_path}/embedding_doc_ids.json", "r") as f:
    doc_ids = json.load(f)

with open(f"{base_path}/doc_clusters.json", "r") as f:
    doc_clusters_json = json.load(f)

doc_embeddings = np.load(f"{base_path}/doc_embeddings.npy")
model = joblib.load(f"{base_path}/embedding_model/model.joblib")

doc_id_to_index = {doc_id: idx for idx, doc_id in enumerate(doc_ids)}

cluster_to_vecs = defaultdict(list)
cluster_to_indices = defaultdict(list)

for doc_id, cluster_id in doc_clusters_json.items():
    idx = doc_id_to_index.get(doc_id)
    if idx is not None:
        cluster_to_vecs[cluster_id].append(doc_embeddings[idx])
        cluster_to_indices[cluster_id].append(idx)

# ÿ≠ÿ≥ÿßÿ® centroids ŸÑŸÉŸÑ ŸÉŸÑÿ≥ÿ™ÿ±
cluster_centroids = {}
for cluster_id, vecs in cluster_to_vecs.items():
    cluster_centroids[int(cluster_id)] = np.mean(vecs, axis=0)

all_cluster_ids = sorted(cluster_centroids.keys())
centroids_matrix = np.array([cluster_centroids[cid] for cid in all_cluster_ids])


stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    tokens = [stemmer.stem(t) for t in tokens]
    return " ".join(tokens)

def search_within_closest_cluster(query, top_k=5):
    cleaned_query = clean_text(query)
    query_vec = model.encode([cleaned_query])  # ŸÑÿßÿ≤ŸÖ ŸäŸÉŸàŸÜ ÿ¥ŸÉŸÑŸáÿß (1, embedding_dim)
    
    cluster_scores = cosine_similarity(query_vec, centroids_matrix)[0]
    closest_cluster_idx = np.argmax(cluster_scores)
    closest_cluster = all_cluster_ids[closest_cluster_idx]
    
    matching_indices = cluster_to_indices[closest_cluster]
    matching_embeddings = doc_embeddings[matching_indices]
    matching_doc_ids = [doc_ids[i] for i in matching_indices]
    
    scores = cosine_similarity(query_vec, matching_embeddings)[0]
    top_indices = scores.argsort()[::-1][:top_k]
    
    results = [{"doc_id": matching_doc_ids[i], "score": float(scores[i])} for i in top_indices]
    return closest_cluster, results

```
