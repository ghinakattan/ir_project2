```python
import pandas as pd
import joblib
import numpy as np
import time

# 1. ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸàÿßŸÑŸÖÿµŸÅŸàŸÅÿßÿ™
doc_ids = pd.read_csv("data/trec_tot/trec_tot_docs.csv")['doc_id'].tolist()

vectorizer = joblib.load("data/trec_tot/tfidf_vectorizer.joblib")
tfidf_matrix = joblib.load("data/trec_tot/tfidf_matrix.joblib")

# 2. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ´ÿßÿ¶ŸÇ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± (5000 ŸÅŸÇÿ∑)
clusters_df = pd.read_csv("data/trec_tot/clusters_df.csv")  # Ÿäÿ≠ÿ™ŸàŸä ÿπŸÑŸâ doc_id ŸÅŸÇÿ∑
cluster_doc_ids = clusters_df['doc_id'].tolist()

# ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑŸÅŸáÿßÿ±ÿ≥ ÿßŸÑŸÖŸÇÿßÿ®ŸÑÿ© ŸÑŸáÿ∞Ÿá ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ŸÖŸÜ ÿßŸÑŸÇÿßÿ¶ŸÖÿ© ÿßŸÑŸÉÿßŸÖŸÑÿ©
doc_id_to_index = {doc_id: idx for idx, doc_id in enumerate(doc_ids)}
cluster_indices = [doc_id_to_index[doc_id] for doc_id in cluster_doc_ids if doc_id in doc_id_to_index]

# ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿ¨ÿ≤ÿ° ŸÖŸÜ tfidf_matrix ŸÅŸÇÿ∑
from scipy.sparse import vstack
tfidf_cluster_subset = vstack([tfidf_matrix[i] for i in cluster_indices])

# 3. ÿßÿ≥ÿ™ÿπŸÑÿßŸÖ ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖ
query_text = "machine learning"
query_vec = vectorizer.transform([query_text])

# 4. ÿßŸÑÿ®ÿ≠ÿ´ ÿßŸÑŸÉÿßŸÖŸÑ
start = time.time()
scores_full = tfidf_matrix.dot(query_vec.T).toarray().ravel()
elapsed_full = time.time() - start
top_k = 10
top_indices_full = scores_full.argsort()[::-1][:top_k]
top_doc_ids_full = [doc_ids[i] for i in top_indices_full]

# 5. ÿßŸÑÿ®ÿ≠ÿ´ ÿØÿßÿÆŸÑ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±
start = time.time()
scores_cluster = tfidf_cluster_subset.dot(query_vec.T).toarray().ravel()
elapsed_cluster = time.time() - start
top_indices_cluster = scores_cluster.argsort()[::-1][:top_k]
top_doc_ids_cluster = [cluster_doc_ids[i] for i in top_indices_cluster]

# 6. ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
print(f"‚è± ÿ≤ŸÖŸÜ ÿßŸÑÿ®ÿ≠ÿ´ ÿßŸÑŸÉÿßŸÖŸÑ: {elapsed_full:.3f} ÿ´ÿßŸÜŸäÿ©")
print("üìÑ ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ (ŸÉÿßŸÖŸÑ):")
for rank, doc_id in enumerate(top_doc_ids_full, 1):
    print(f"{rank}. doc_id: {doc_id}")

print("\n‚è± ÿ≤ŸÖŸÜ ÿßŸÑÿ®ÿ≠ÿ´ ÿØÿßÿÆŸÑ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±: {:.3f} ÿ´ÿßŸÜŸäÿ©".format(elapsed_cluster))
print("üìÑ ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ (ÿØÿßÿÆŸÑ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±):")
for rank, doc_id in enumerate(top_doc_ids_cluster, 1):
    print(f"{rank}. doc_id: {doc_id}")

# 7. ŸÖŸÇÿßÿ±ŸÜÿ© ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
print("\nüîÅ ŸÖŸÇÿßÿ±ŸÜÿ© ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ:")
for i in range(top_k):
    doc_full = top_doc_ids_full[i] if i < len(top_doc_ids_full) else "N/A"
    doc_cluster = top_doc_ids_cluster[i] if i < len(top_doc_ids_cluster) else "N/A"
    print(f"{i+1}. Full: {doc_full} | Cluster: {doc_cluster}")
```

```python
import pandas as pd
import joblib
import numpy as np
import time
from scipy.sparse import load_npz, vstack

# 1. ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸàÿßŸÑŸÖÿµŸÅŸàŸÅÿßÿ™
doc_ids_path = "data/trec_tot/trec_tot_docs.csv"
vectorizer_path = "data/trec_tot/tfidf_vectorizer.pkl"
tfidf_matrix_path = "data/trec_tot/tfidf_docs_matrix.npz"
cluster_df_path = "data/trec_tot/tfidf_clusters.csv"

doc_ids = pd.read_csv(doc_ids_path)['doc_id'].tolist()
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

# 2. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ´ÿßÿ¶ŸÇ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± (5000 ŸÅŸÇÿ∑)
clusters_df = pd.read_csv(cluster_df_path)  # Ÿäÿ≠ÿ™ŸàŸä ÿπŸÑŸâ doc_id ŸÅŸÇÿ∑
cluster_doc_ids = clusters_df['doc_id'].tolist()

# ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑŸÅŸáÿßÿ±ÿ≥ ÿßŸÑŸÖŸÇÿßÿ®ŸÑÿ© ŸÑŸáÿ∞Ÿá ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ŸÖŸÜ ÿßŸÑŸÇÿßÿ¶ŸÖÿ© ÿßŸÑŸÉÿßŸÖŸÑÿ©
doc_id_to_index = {doc_id: idx for idx, doc_id in enumerate(doc_ids)}
cluster_indices = [doc_id_to_index[doc_id] for doc_id in cluster_doc_ids if doc_id in doc_id_to_index]

# ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿ¨ÿ≤ÿ° ŸÖŸÜ tfidf_matrix ŸÅŸÇÿ∑
tfidf_cluster_subset = vstack([tfidf_matrix[i] for i in cluster_indices])

# 3. ÿßÿ≥ÿ™ÿπŸÑÿßŸÖ ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖ
query_text = "machine learning"
query_vec = vectorizer.transform([query_text])

# 4. ÿßŸÑÿ®ÿ≠ÿ´ ÿßŸÑŸÉÿßŸÖŸÑ
start = time.time()
scores_full = tfidf_matrix.dot(query_vec.T).toarray().ravel()
elapsed_full = time.time() - start
top_k = 10
top_indices_full = scores_full.argsort()[::-1][:top_k]
top_doc_ids_full = [doc_ids[i] for i in top_indices_full]

# 5. ÿßŸÑÿ®ÿ≠ÿ´ ÿØÿßÿÆŸÑ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±
start = time.time()
scores_cluster = tfidf_cluster_subset.dot(query_vec.T).toarray().ravel()
elapsed_cluster = time.time() - start
top_indices_cluster = scores_cluster.argsort()[::-1][:top_k]
top_doc_ids_cluster = [cluster_doc_ids[i] for i in top_indices_cluster]

# 6. ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
print(f"‚è± ÿ≤ŸÖŸÜ ÿßŸÑÿ®ÿ≠ÿ´ ÿßŸÑŸÉÿßŸÖŸÑ: {elapsed_full:.3f} ÿ´ÿßŸÜŸäÿ©")
print("üìÑ ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ (ŸÉÿßŸÖŸÑ):")
for rank, doc_id in enumerate(top_doc_ids_full, 1):
    print(f"{rank}. doc_id: {doc_id}")

print("\n‚è± ÿ≤ŸÖŸÜ ÿßŸÑÿ®ÿ≠ÿ´ ÿØÿßÿÆŸÑ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±: {:.3f} ÿ´ÿßŸÜŸäÿ©".format(elapsed_cluster))
print("üìÑ ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ (ÿØÿßÿÆŸÑ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±):")
for rank, doc_id in enumerate(top_doc_ids_cluster, 1):
    print(f"{rank}. doc_id: {doc_id}")

# 7. ŸÖŸÇÿßÿ±ŸÜÿ© ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
print("\nüîÅ ŸÖŸÇÿßÿ±ŸÜÿ© ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ:")
for i in range(top_k):
    doc_full = top_doc_ids_full[i] if i < len(top_doc_ids_full) else "N/A"
    doc_cluster = top_doc_ids_cluster[i] if i < len(top_doc_ids_cluster) else "N/A"
    print(f"{i+1}. Full: {doc_full} | Cluster: {doc_cluster}")
```

```python
import pickle
import joblib
import numpy as np
import pandas as pd
import time
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import load_npz
from sklearn.metrics.pairwise import cosine_similarity

# 1. ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ŸàÿßŸÑŸÖÿµŸÅŸàŸÅÿ©
tfidf_matrix = load_npz("data/trec_tot/tfidf_docs_matrix.npz")

# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÄ vectorizer ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ joblib ŸÖÿ®ÿßÿ¥ÿ±ÿ©
vectorizer = joblib.load("data/trec_tot/tfidf_vectorizer.pkl")

# 2. ÿ™ÿ≠ŸÖŸäŸÑ ŸÖŸÑŸÅ doc_ids (ÿßŸÑÿ™ÿ±ÿ™Ÿäÿ® Ÿäÿ¨ÿ® ÿ£ŸÜ Ÿäÿ™ÿ∑ÿßÿ®ŸÇ ŸÖÿπ tfidf_matrix)
doc_ids_df = pd.read_csv("data/trec_tot/trec_tot_docs.csv")
doc_ids = doc_ids_df['doc_id'].tolist()

# 3. ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± ŸÑŸÉŸÑ Ÿàÿ´ŸäŸÇÿ©
clusters_df = pd.read_csv("data/trec_tot/tfidf_clusters.csv")  # ŸÖÿ´ÿßŸÑ
cluster_labels = clusters_df['cluster_label'].values

# 4. ÿ≠ÿ≥ÿßÿ® ŸÖÿ±ÿßŸÉÿ≤ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± (centroids)
num_clusters = max(cluster_labels) + 1
centroids = []

for c in range(num_clusters):
    idx = np.where(cluster_labels == c)[0]
    cluster_vectors = tfidf_matrix[idx]
    # ÿ≠ÿ≥ÿßÿ® ÿßŸÑŸÖÿ™Ÿàÿ≥ÿ∑ Ÿàÿ™ÿ≠ŸàŸäŸÑŸá ÿ•ŸÑŸâ numpy array (1D)
    centroid = cluster_vectors.mean(axis=0).A1  
    centroids.append(centroid)

centroids = np.vstack(centroids)  # ŸÖÿµŸÅŸàŸÅÿ© numpy (num_clusters x num_features)

# 5. ÿßÿ≥ÿ™ÿπŸÑÿßŸÖ ÿßŸÑŸÖÿ≥ÿ™ÿÆÿØŸÖ Ÿàÿ™ŸÖÿ´ŸäŸÑŸá
query_text = "machine learning"
query_vec = vectorizer.transform([query_text])

# 6. ÿ™ÿ≠ÿØŸäÿØ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± ÿßŸÑÿ£ŸÇÿ±ÿ® ŸÑŸÑÿßÿ≥ÿ™ÿπŸÑÿßŸÖ
similarities = cosine_similarity(query_vec, centroids)
best_cluster = similarities.argmax()

print(f"ÿ£ŸÅÿ∂ŸÑ ŸÉŸÑÿßÿ≥ÿ™ÿ± ŸÑŸÑÿßÿ≥ÿ™ÿπŸÑÿßŸÖ: {best_cluster} (ÿßŸÑÿ™ÿ¥ÿßÿ®Ÿá: {similarities[0, best_cluster]:.4f})")

# 7. ÿßŸÑÿ®ÿ≠ÿ´ ÿØÿßÿÆŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ÿßŸÑÿ™Ÿä ÿ™ŸÜÿ™ŸÖŸä ŸÑŸáÿ∞ÿß ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± ŸÅŸÇÿ∑
indices_in_cluster = np.where(cluster_labels == best_cluster)[0]
tfidf_cluster_subset = tfidf_matrix[indices_in_cluster]

start = time.time()
scores = tfidf_cluster_subset.dot(query_vec.T).toarray().ravel()
elapsed = time.time() - start

top_k = 10
top_indices = scores.argsort()[::-1][:top_k]
top_doc_ids = [doc_ids[indices_in_cluster[i]] for i in top_indices]

print(f"‚è± ÿ≤ŸÖŸÜ ÿßŸÑÿ®ÿ≠ÿ´ ÿØÿßÿÆŸÑ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±: {elapsed:.3f} ÿ´ÿßŸÜŸäÿ©")
print("üìÑ ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ÿØÿßÿÆŸÑ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±:")
for rank, doc_id in enumerate(top_doc_ids, 1):
    print(f"{rank}. doc_id: {doc_id}")

# 8. (ÿßÿÆÿ™Ÿäÿßÿ±Ÿä) ÿßŸÑÿ®ÿ≠ÿ´ ÿßŸÑŸÉÿßŸÖŸÑ ŸÑŸÑŸÖŸÇÿßÿ±ŸÜÿ©
start_full = time.time()
scores_full = tfidf_matrix.dot(query_vec.T).toarray().ravel()
elapsed_full = time.time() - start_full

top_indices_full = scores_full.argsort()[::-1][:top_k]
top_doc_ids_full = [doc_ids[i] for i in top_indices_full]

print(f"\n‚è± ÿ≤ŸÖŸÜ ÿßŸÑÿ®ÿ≠ÿ´ ÿßŸÑŸÉÿßŸÖŸÑ: {elapsed_full:.3f} ÿ´ÿßŸÜŸäÿ©")
print("üìÑ ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ (ŸÉÿßŸÖŸÑ):")
for rank, doc_id in enumerate(top_doc_ids_full, 1):
    print(f"{rank}. doc_id: {doc_id}")

# 9. ŸÖŸÇÿßÿ±ŸÜÿ© ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
print("\nüîÅ ŸÖŸÇÿßÿ±ŸÜÿ© ÿ£ŸÅÿ∂ŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ:")
for i in range(top_k):
    print(f"{i+1}. Full: {top_doc_ids_full[i]} | Cluster: {top_doc_ids[i]}")
```

```python
import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿäÿ© ===
dataset = "trec_tot"  # ÿ£Ÿà antique
project_root = os.getcwd()
base_path = os.path.join(project_root, 'data', dataset)
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")
doc_file = os.path.join(base_path, "tot_docs_clean.csv")  # ‚úÖ ÿ™ŸÖ ÿ™ÿπÿØŸäŸÑŸá ÿ≠ÿ≥ÿ® ÿ∑ŸÑÿ®ŸÉ

# === ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ===
print("üì¶ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸàÿßŸÑŸÖÿµŸÅŸàŸÅÿ©...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"‚úÖ ÿ¥ŸÉŸÑ ÿßŸÑŸÖÿµŸÅŸàŸÅÿ©: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

# === ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP ===
print("üìâ ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP...")
umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]
reduced_data = umap.fit_transform(sample_matrix)

# === ÿ™ÿ¨ÿ±ÿ®ÿ© ÿ£ÿπÿØÿßÿØ ŸÖÿÆÿ™ŸÑŸÅÿ© ŸÖŸÜ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±ÿßÿ™ ===
cluster_range = [5, 10, 20, 50, 100]
results = []

for n_clusters in cluster_range:
    print(f"\nüîÑ ÿ™ÿ¨ÿ±Ÿäÿ® ÿπÿØÿØ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±ÿßÿ™: {n_clusters}")
    kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
    labels = kmeans.fit_predict(reduced_data)
    
    sil_score = silhouette_score(reduced_data, labels)
    db_score = davies_bouldin_score(reduced_data, labels)
    ch_score = calinski_harabasz_score(reduced_data, labels)
    
    print(f"Silhouette Score: {sil_score:.3f}")
    print(f"Davies-Bouldin Index: {db_score:.3f} (ŸÉŸÑŸÖÿß ÿ£ŸÇŸÑ ÿ£ŸÅÿ∂ŸÑ)")
    print(f"Calinski-Harabasz Index: {ch_score:.3f}")
    
    results.append({
        "n_clusters": n_clusters,
        "silhouette": sil_score,
        "davies_bouldin": db_score,
        "calinski_harabasz": ch_score,
        "labels": labels,          # ŸÑŸà ÿ≠ÿ®Ÿäÿ™ ÿ™ÿ≠ÿ™ŸÅÿ∏ ÿ®ÿßŸÑÿ™ÿµŸÜŸäŸÅÿßÿ™
        "kmeans_model": kmeans     # ŸÑŸà ÿ≠ÿ®Ÿäÿ™ ÿ™ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸÑÿßÿ≠ŸÇÿßŸã
    })

# === ÿ±ÿ≥ŸÖ ŸÖÿ§ÿ¥ÿ±ÿßÿ™ ÿßŸÑÿ™ŸÇŸäŸäŸÖ ŸÖŸÇÿßÿ®ŸÑ ÿπÿØÿØ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±ÿßÿ™ ===
df_results = pd.DataFrame(results)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(df_results["n_clusters"], df_results["silhouette"], marker='o')
plt.title("Silhouette Score")
plt.xlabel("Number of Clusters")
plt.ylabel("Score")
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(df_results["n_clusters"], df_results["davies_bouldin"], marker='o', color='red')
plt.title("Davies-Bouldin Index (ŸÉŸÑŸÖÿß ÿ£ŸÇŸÑ ÿ£ŸÅÿ∂ŸÑ)")
plt.xlabel("Number of Clusters")
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(df_results["n_clusters"], df_results["calinski_harabasz"], marker='o', color='green')
plt.title("Calinski-Harabasz Index")
plt.xlabel("Number of Clusters")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import re
import os
import joblib
import numpy as np
import pandas as pd
from scipy.sparse import save_npz
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
import nltk

# ÿ™ÿ≠ŸÖŸäŸÑ ŸÖŸàÿßÿ±ÿØ NLTK
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")


stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def custom_preprocessor(text):
    if text is None:
        return ""
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text

def custom_tokenizer(text):
    text = custom_preprocessor(text)
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    tokens = [stemmer.stem(t) for t in tokens]
    return tokens


dataset = "antique"

base_path = os.path.join("data", dataset)
os.makedirs(base_path, exist_ok=True)

doc_file = os.path.join(base_path, "antique_docs_clean.csv") if dataset == "antique" else os.path.join(base_path, "tot_docs_clean.csv")
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")



df = pd.read_csv(doc_file)
texts = df["doc_text"].astype(str).tolist()

vectorizer = TfidfVectorizer(
    preprocessor=custom_preprocessor,
    tokenizer=custom_tokenizer,
    lowercase=False,
    max_df=0.9,
    min_df=5
)

tfidf_matrix = vectorizer.fit_transform(texts)

joblib.dump(vectorizer, vectorizer_path)
save_npz(matrix_path, tfidf_matrix)

print("‚úÖ ÿ™ŸÖ ÿ≠ŸÅÿ∏ vectorizer Ÿà tfidf_matrix ÿ®ŸÜÿ¨ÿßÿ≠")



import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿäÿ© ===
dataset = "antique"  # ÿ£Ÿà antique
project_root = os.getcwd()
base_path = os.path.join(project_root, 'data', dataset)
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")
doc_file = os.path.join(base_path, "antique_docs_clean.csv")  # ‚úÖ ÿ™ŸÖ ÿ™ÿπÿØŸäŸÑŸá ÿ≠ÿ≥ÿ® ÿ∑ŸÑÿ®ŸÉ

# === ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ===
print("üì¶ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸàÿßŸÑŸÖÿµŸÅŸàŸÅÿ©...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"‚úÖ ÿ¥ŸÉŸÑ ÿßŸÑŸÖÿµŸÅŸàŸÅÿ©: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

# === ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP ===
print("üìâ ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP...")
umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]
reduced_data = umap.fit_transform(sample_matrix)

# === ÿ™ÿ¨ÿ±ÿ®ÿ© ÿ£ÿπÿØÿßÿØ ŸÖÿÆÿ™ŸÑŸÅÿ© ŸÖŸÜ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±ÿßÿ™ ===
cluster_range = [5, 10, 20, 50, 100]
results = []

for n_clusters in cluster_range:
    print(f"\nüîÑ ÿ™ÿ¨ÿ±Ÿäÿ® ÿπÿØÿØ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±ÿßÿ™: {n_clusters}")
    kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
    labels = kmeans.fit_predict(reduced_data)
    
    sil_score = silhouette_score(reduced_data, labels)
    db_score = davies_bouldin_score(reduced_data, labels)
    ch_score = calinski_harabasz_score(reduced_data, labels)
    
    print(f"Silhouette Score: {sil_score:.3f}")
    print(f"Davies-Bouldin Index: {db_score:.3f} (ŸÉŸÑŸÖÿß ÿ£ŸÇŸÑ ÿ£ŸÅÿ∂ŸÑ)")
    print(f"Calinski-Harabasz Index: {ch_score:.3f}")
    
    results.append({
        "n_clusters": n_clusters,
        "silhouette": sil_score,
        "davies_bouldin": db_score,
        "calinski_harabasz": ch_score,
        "labels": labels,          # ŸÑŸà ÿ≠ÿ®Ÿäÿ™ ÿ™ÿ≠ÿ™ŸÅÿ∏ ÿ®ÿßŸÑÿ™ÿµŸÜŸäŸÅÿßÿ™
        "kmeans_model": kmeans     # ŸÑŸà ÿ≠ÿ®Ÿäÿ™ ÿ™ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸÑÿßÿ≠ŸÇÿßŸã
    })

# === ÿ±ÿ≥ŸÖ ŸÖÿ§ÿ¥ÿ±ÿßÿ™ ÿßŸÑÿ™ŸÇŸäŸäŸÖ ŸÖŸÇÿßÿ®ŸÑ ÿπÿØÿØ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ±ÿßÿ™ ===
df_results = pd.DataFrame(results)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(df_results["n_clusters"], df_results["silhouette"], marker='o')
plt.title("Silhouette Score")
plt.xlabel("Number of Clusters")
plt.ylabel("Score")
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(df_results["n_clusters"], df_results["davies_bouldin"], marker='o', color='red')
plt.title("Davies-Bouldin Index (ŸÉŸÑŸÖÿß ÿ£ŸÇŸÑ ÿ£ŸÅÿ∂ŸÑ)")
plt.xlabel("Number of Clusters")
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(df_results["n_clusters"], df_results["calinski_harabasz"], marker='o', color='green')
plt.title("Calinski-Harabasz Index")
plt.xlabel("Number of Clusters")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿäÿ© ===
dataset = "antique"  # ÿ£Ÿà antique
project_root = os.getcwd()
base_path = os.path.join(project_root, 'data', dataset)
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")
doc_file = os.path.join(base_path, "antique_docs_clean.csv")  # ‚úÖ ÿ™ŸÖ ÿ™ÿπÿØŸäŸÑŸá ÿ≠ÿ≥ÿ® ÿ∑ŸÑÿ®ŸÉ

# === ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ===
print("üì¶ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸàÿßŸÑŸÖÿµŸÅŸàŸÅÿ©...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"‚úÖ ÿ¥ŸÉŸÑ ÿßŸÑŸÖÿµŸÅŸàŸÅÿ©: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

# === ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP ===
print("üìâ ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP...")
umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]
reduced_data = umap.fit_transform(sample_matrix)

# === ÿπŸÜŸÇÿØÿ© ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ KMeans ===
print("üîÑ ÿ™ŸÜŸÅŸäÿ∞ MiniBatchKMeans...")
n_clusters = 5
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
labels = kmeans.fit_predict(reduced_data)

# === ÿπÿ±ÿ∂ ÿßŸÑŸÉŸÑŸÖÿßÿ™ ÿßŸÑŸÖŸÅÿ™ÿßÿ≠Ÿäÿ© ŸÑŸÉŸÑ ÿπŸÜŸÇŸàÿØ ===
print("\nüìå ÿ£ŸáŸÖ ÿßŸÑŸÉŸÑŸÖÿßÿ™ ŸÑŸÉŸÑ ÿπŸÜŸÇŸàÿØ:")
terms = vectorizer.get_feature_names_out()
for i in range(n_clusters):
    print(f"\nCluster {i}:")
    cluster_indices = np.where(labels == i)[0]
    cluster_matrix = sample_matrix[cluster_indices].mean(axis=0).A1
    top_indices = cluster_matrix.argsort()[::-1][:10]
    top_terms = [terms[j] for j in top_indices]
    print(", ".join(top_terms))

# === ÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿπŸÜŸÇÿØÿ© ===
print("\nüìà ÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿπŸÜŸÇÿØÿ©:")
print(f"Silhouette Score: {silhouette_score(reduced_data, labels):.3f}")
print(f"Davies-Bouldin Index: {davies_bouldin_score(reduced_data, labels):.3f}")
print(f"Calinski-Harabasz Index: {calinski_harabasz_score(reduced_data, labels):.3f}")

# === ÿ±ÿ≥ŸÖ ÿßŸÑÿπŸÜŸÇŸàÿØ ===
print("üìä ÿ±ÿ≥ŸÖ ÿßŸÑÿπŸÜŸÇŸàÿØ...")
plt.figure(figsize=(10, 7))
colors = plt.cm.get_cmap('tab10', n_clusters)
for i in range(n_clusters):
    cluster_points = reduced_data[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                label=f"Cluster {i}", s=30, alpha=0.6, color=colors(i))

plt.title("üîµ ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ÿ≠ÿ≥ÿ® ÿßŸÑÿπŸÜŸÇŸàÿØ")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend()
plt.grid(True)
plt.tight_layout()

plot_path = os.path.join(base_path, "clusters_plot.png")
plt.savefig(plot_path)
plt.show()

# === ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ŸÉŸÄ CSV ===
print("\nüíæ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ...")
docs_df = pd.DataFrame({"doc_id": np.array(doc_ids)[sample_indices]})
docs_df["cluster_label"] = labels
results_path = os.path.join(base_path, "tfidf_clusters.csv")
docs_df.to_csv(results_path, index=False)
print(f"üìÅ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ŸÅŸä: {results_path}")

# === ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ UMAP ===
umap_model_path = os.path.join(base_path, "umap_model.joblib")
joblib.dump(umap, umap_model_path)
print(f"üíæ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ UMAP ŸÅŸä: {umap_model_path}")

# === ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± KMeans ===
cluster_model_path = os.path.join(base_path, "cluster_model.joblib")
joblib.dump(kmeans, cluster_model_path)
print(f"üíæ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ KMeans ŸÅŸä: {cluster_model_path}")

print(f"\n‚úÖ ÿßŸÑÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÉÿ™ŸÖŸÑ. ÿ™ŸÖ ÿ≠ŸÅÿ∏ ÿßŸÑŸÖÿÆÿ∑ÿ∑ ŸÅŸä '{plot_path}'")
```

```python
import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿäÿ© ===
dataset = "antique"  # ÿ£Ÿà "antique"
project_root = os.getcwd()

# ÿ≠ÿ≥ÿ® ÿßŸÑŸÖÿ≥ÿßÿ±ÿßÿ™ ÿßŸÑÿ™Ÿä ÿ£ÿ±ÿ≥ŸÑÿ™Ÿáÿßÿå ŸÖŸÑŸÅÿßÿ™ tfidf ÿ∂ŸÖŸÜ offline_indexing_service
base_offline_indexing_path = os.path.join(project_root, "offline_indexing_service", "data", dataset)

vectorizer_path = os.path.join(base_offline_indexing_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_offline_indexing_path, "tfidf_docs_matrix.npz")

# ŸÖŸÑŸÅÿßÿ™ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ÿßŸÑÿ£ÿµŸÑŸäÿ© ŸàÿßŸÑŸÖŸÜÿ∏ŸÅÿ© ÿ∂ŸÖŸÜ ŸÖÿ¨ŸÑÿØ data ŸÖÿ®ÿßÿ¥ÿ±ÿ© (ŸàŸÑŸäÿ≥ ÿ∂ŸÖŸÜ offline_indexing_service)
base_data_path = os.path.join(project_root, "data", dataset)

if dataset == "trec_tot":
    doc_file = os.path.join(base_data_path, "trec_tot_docs_clean.csv")
elif dataset == "antique":
    doc_file = os.path.join(base_data_path, "antique_docs_clean.csv")
else:
    doc_file = None  # ÿ£Ÿà ÿ±ŸÅÿπ ÿßÿ≥ÿ™ÿ´ŸÜÿßÿ° ŸáŸÜÿß

print("vectorizer_path:", vectorizer_path)
print("tfidf_matrix_path:", tfidf_matrix_path)
print("doc_file:", doc_file)

# === ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ===
print("üì¶ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸàÿßŸÑŸÖÿµŸÅŸàŸÅÿ©...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"‚úÖ ÿ¥ŸÉŸÑ ÿßŸÑŸÖÿµŸÅŸàŸÅÿ©: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

# === ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP ===
print("üìâ ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP...")
umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]
reduced_data = umap.fit_transform(sample_matrix)

# === ÿπŸÜŸÇÿØÿ© ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ KMeans ===
print("üîÑ ÿ™ŸÜŸÅŸäÿ∞ MiniBatchKMeans...")
n_clusters = 5
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
labels = kmeans.fit_predict(reduced_data)

# === ÿπÿ±ÿ∂ ÿßŸÑŸÉŸÑŸÖÿßÿ™ ÿßŸÑŸÖŸÅÿ™ÿßÿ≠Ÿäÿ© ŸÑŸÉŸÑ ÿπŸÜŸÇŸàÿØ ===
print("\nüìå ÿ£ŸáŸÖ ÿßŸÑŸÉŸÑŸÖÿßÿ™ ŸÑŸÉŸÑ ÿπŸÜŸÇŸàÿØ:")
terms = vectorizer.get_feature_names_out()
for i in range(n_clusters):
    print(f"\nCluster {i}:")
    cluster_indices = np.where(labels == i)[0]
    cluster_matrix = sample_matrix[cluster_indices].mean(axis=0).A1
    top_indices = cluster_matrix.argsort()[::-1][:10]
    top_terms = [terms[j] for j in top_indices]
    print(", ".join(top_terms))

# === ÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿπŸÜŸÇÿØÿ© ===
print("\nüìà ÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿπŸÜŸÇÿØÿ©:")
print(f"Silhouette Score: {silhouette_score(reduced_data, labels):.3f}")
print(f"Davies-Bouldin Index: {davies_bouldin_score(reduced_data, labels):.3f}")
print(f"Calinski-Harabasz Index: {calinski_harabasz_score(reduced_data, labels):.3f}")

# === ÿ±ÿ≥ŸÖ ÿßŸÑÿπŸÜŸÇŸàÿØ ===
print("üìä ÿ±ÿ≥ŸÖ ÿßŸÑÿπŸÜŸÇŸàÿØ...")
plt.figure(figsize=(10, 7))
colors = plt.cm.get_cmap('tab10', n_clusters)
for i in range(n_clusters):
    cluster_points = reduced_data[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                label=f"Cluster {i}", s=30, alpha=0.6, color=colors(i))

plt.title("üîµ ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ÿ≠ÿ≥ÿ® ÿßŸÑÿπŸÜŸÇŸàÿØ")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend()
plt.grid(True)
plt.tight_layout()

plot_path = os.path.join(base_offline_indexing_path, "clusters_plot.png")
plt.savefig(plot_path)
plt.show()

# === ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ŸÉŸÄ CSV ===
print("\nüíæ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ...")
docs_df = pd.DataFrame({"doc_id": np.array(doc_ids)[sample_indices]})
docs_df["cluster_label"] = labels
results_path = os.path.join(base_offline_indexing_path, "tfidf_clusters.csv")
docs_df.to_csv(results_path, index=False)
print(f"üìÅ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ŸÅŸä: {results_path}")

# === ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ UMAP ===
umap_model_path = os.path.join(base_offline_indexing_path, "umap_model.joblib")
joblib.dump(umap, umap_model_path)
print(f"üíæ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ UMAP ŸÅŸä: {umap_model_path}")

# === ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± KMeans ===
cluster_model_path = os.path.join(base_offline_indexing_path, "cluster_model.joblib")
joblib.dump(kmeans, cluster_model_path)
print(f"üíæ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ KMeans ŸÅŸä: {cluster_model_path}")

print(f"\n‚úÖ ÿßŸÑÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÉÿ™ŸÖŸÑ. ÿ™ŸÖ ÿ≠ŸÅÿ∏ ÿßŸÑŸÖÿÆÿ∑ÿ∑ ŸÅŸä '{plot_path}'")
```
