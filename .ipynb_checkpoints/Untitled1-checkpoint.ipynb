```python
import pickle
import joblib
import numpy as np
import pandas as pd
import time
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import load_npz
from sklearn.metrics.pairwise import cosine_similarity

# 1. تحميل بيانات الوثائق والمصفوفة
tfidf_matrix = load_npz("data/trec_tot/tfidf_docs_matrix.npz")

# تحميل الـ vectorizer باستخدام joblib مباشرة
vectorizer = joblib.load("data/trec_tot/tfidf_vectorizer.pkl")

# 2. تحميل ملف doc_ids (الترتيب يجب أن يتطابق مع tfidf_matrix)
doc_ids_df = pd.read_csv("data/trec_tot/trec_tot_docs.csv")
doc_ids = doc_ids_df['doc_id'].tolist()

# 3. تحميل بيانات الكلاستر لكل وثيقة
clusters_df = pd.read_csv("data/trec_tot/tfidf_clusters.csv")  # مثال
cluster_labels = clusters_df['cluster_label'].values

# 4. حساب مراكز الكلاستر (centroids)
num_clusters = max(cluster_labels) + 1
centroids = []

for c in range(num_clusters):
    idx = np.where(cluster_labels == c)[0]
    cluster_vectors = tfidf_matrix[idx]
    # حساب المتوسط وتحويله إلى numpy array (1D)
    centroid = cluster_vectors.mean(axis=0).A1  
    centroids.append(centroid)

centroids = np.vstack(centroids)  # مصفوفة numpy (num_clusters x num_features)

# 5. استعلام المستخدم وتمثيله
query_text = "machine learning"
query_vec = vectorizer.transform([query_text])

# 6. تحديد الكلاستر الأقرب للاستعلام
similarities = cosine_similarity(query_vec, centroids)
best_cluster = similarities.argmax()

print(f"أفضل كلاستر للاستعلام: {best_cluster} (التشابه: {similarities[0, best_cluster]:.4f})")

# 7. البحث داخل الوثائق التي تنتمي لهذا الكلاستر فقط
indices_in_cluster = np.where(cluster_labels == best_cluster)[0]
tfidf_cluster_subset = tfidf_matrix[indices_in_cluster]

start = time.time()
scores = tfidf_cluster_subset.dot(query_vec.T).toarray().ravel()
elapsed = time.time() - start

top_k = 10
top_indices = scores.argsort()[::-1][:top_k]
top_doc_ids = [doc_ids[indices_in_cluster[i]] for i in top_indices]

print(f"⏱ زمن البحث داخل الكلاستر: {elapsed:.3f} ثانية")
print("📄 أفضل الوثائق داخل الكلاستر:")
for rank, doc_id in enumerate(top_doc_ids, 1):
    print(f"{rank}. doc_id: {doc_id}")

# 8. (اختياري) البحث الكامل للمقارنة
start_full = time.time()
scores_full = tfidf_matrix.dot(query_vec.T).toarray().ravel()
elapsed_full = time.time() - start_full

top_indices_full = scores_full.argsort()[::-1][:top_k]
top_doc_ids_full = [doc_ids[i] for i in top_indices_full]

print(f"\n⏱ زمن البحث الكامل: {elapsed_full:.3f} ثانية")
print("📄 أفضل الوثائق (كامل):")
for rank, doc_id in enumerate(top_doc_ids_full, 1):
    print(f"{rank}. doc_id: {doc_id}")

# 9. مقارنة النتائج
print("\n🔁 مقارنة أفضل الوثائق:")
for i in range(top_k):
    print(f"{i+1}. Full: {top_doc_ids_full[i]} | Cluster: {top_doc_ids[i]}")
```

```python
import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === الإعدادات الأساسية ===
dataset = "trec_tot"  # أو antique
project_root = os.getcwd()
base_path = os.path.join(project_root, 'data', dataset)
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")
doc_file = os.path.join(base_path, "tot_docs_clean.csv")  # ✅ تم تعديله حسب طلبك

# === تحميل البيانات ===
print("📦 تحميل النموذج والمصفوفة...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"✅ شكل المصفوفة: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

# === تقليل الأبعاد باستخدام UMAP ===
print("📉 تقليل الأبعاد باستخدام UMAP...")
umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]
reduced_data = umap.fit_transform(sample_matrix)

# === تجربة أعداد مختلفة من الكلاسترات ===
cluster_range = [5, 10, 20, 50, 100]
results = []

for n_clusters in cluster_range:
    print(f"\n🔄 تجريب عدد الكلاسترات: {n_clusters}")
    kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
    labels = kmeans.fit_predict(reduced_data)
    
    sil_score = silhouette_score(reduced_data, labels)
    db_score = davies_bouldin_score(reduced_data, labels)
    ch_score = calinski_harabasz_score(reduced_data, labels)
    
    print(f"Silhouette Score: {sil_score:.3f}")
    print(f"Davies-Bouldin Index: {db_score:.3f} (كلما أقل أفضل)")
    print(f"Calinski-Harabasz Index: {ch_score:.3f}")
    
    results.append({
        "n_clusters": n_clusters,
        "silhouette": sil_score,
        "davies_bouldin": db_score,
        "calinski_harabasz": ch_score,
        "labels": labels,          # لو حبيت تحتفظ بالتصنيفات
        "kmeans_model": kmeans     # لو حبيت تحفظ النموذج لاحقاً
    })

# === رسم مؤشرات التقييم مقابل عدد الكلاسترات ===
df_results = pd.DataFrame(results)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(df_results["n_clusters"], df_results["silhouette"], marker='o')
plt.title("Silhouette Score")
plt.xlabel("Number of Clusters")
plt.ylabel("Score")
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(df_results["n_clusters"], df_results["davies_bouldin"], marker='o', color='red')
plt.title("Davies-Bouldin Index (كلما أقل أفضل)")
plt.xlabel("Number of Clusters")
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(df_results["n_clusters"], df_results["calinski_harabasz"], marker='o', color='green')
plt.title("Calinski-Harabasz Index")
plt.xlabel("Number of Clusters")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import re
import os
import joblib
import numpy as np
import pandas as pd
from scipy.sparse import save_npz
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
import nltk

# تحميل موارد NLTK
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")


stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def custom_preprocessor(text):
    if text is None:
        return ""
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text

def custom_tokenizer(text):
    text = custom_preprocessor(text)
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    tokens = [stemmer.stem(t) for t in tokens]
    return tokens


dataset = "antique"

base_path = os.path.join("data", dataset)
os.makedirs(base_path, exist_ok=True)

doc_file = os.path.join(base_path, "antique_docs_clean.csv") if dataset == "antique" else os.path.join(base_path, "tot_docs_clean.csv")
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")



df = pd.read_csv(doc_file)
texts = df["doc_text"].astype(str).tolist()

vectorizer = TfidfVectorizer(
    preprocessor=custom_preprocessor,
    tokenizer=custom_tokenizer,
    lowercase=False,
    max_df=0.9,
    min_df=5
)

tfidf_matrix = vectorizer.fit_transform(texts)

joblib.dump(vectorizer, vectorizer_path)
save_npz(matrix_path, tfidf_matrix)

print("✅ تم حفظ vectorizer و tfidf_matrix بنجاح")



import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === الإعدادات الأساسية ===
dataset = "antique"  # أو antique
project_root = os.getcwd()
base_path = os.path.join(project_root, 'data', dataset)
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")
doc_file = os.path.join(base_path, "antique_docs_clean.csv")  # ✅ تم تعديله حسب طلبك

# === تحميل البيانات ===
print("📦 تحميل النموذج والمصفوفة...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"✅ شكل المصفوفة: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

# === تقليل الأبعاد باستخدام UMAP ===
print("📉 تقليل الأبعاد باستخدام UMAP...")
umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]
reduced_data = umap.fit_transform(sample_matrix)

# === تجربة أعداد مختلفة من الكلاسترات ===
cluster_range = [5, 10, 20, 50, 100]
results = []

for n_clusters in cluster_range:
    print(f"\n🔄 تجريب عدد الكلاسترات: {n_clusters}")
    kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
    labels = kmeans.fit_predict(reduced_data)
    
    sil_score = silhouette_score(reduced_data, labels)
    db_score = davies_bouldin_score(reduced_data, labels)
    ch_score = calinski_harabasz_score(reduced_data, labels)
    
    print(f"Silhouette Score: {sil_score:.3f}")
    print(f"Davies-Bouldin Index: {db_score:.3f} (كلما أقل أفضل)")
    print(f"Calinski-Harabasz Index: {ch_score:.3f}")
    
    results.append({
        "n_clusters": n_clusters,
        "silhouette": sil_score,
        "davies_bouldin": db_score,
        "calinski_harabasz": ch_score,
        "labels": labels,          # لو حبيت تحتفظ بالتصنيفات
        "kmeans_model": kmeans     # لو حبيت تحفظ النموذج لاحقاً
    })

# === رسم مؤشرات التقييم مقابل عدد الكلاسترات ===
df_results = pd.DataFrame(results)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(df_results["n_clusters"], df_results["silhouette"], marker='o')
plt.title("Silhouette Score")
plt.xlabel("Number of Clusters")
plt.ylabel("Score")
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(df_results["n_clusters"], df_results["davies_bouldin"], marker='o', color='red')
plt.title("Davies-Bouldin Index (كلما أقل أفضل)")
plt.xlabel("Number of Clusters")
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(df_results["n_clusters"], df_results["calinski_harabasz"], marker='o', color='green')
plt.title("Calinski-Harabasz Index")
plt.xlabel("Number of Clusters")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === الإعدادات الأساسية ===
dataset = "antique"  # أو antique
project_root = os.getcwd()
base_path = os.path.join(project_root, 'data', dataset)
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")
doc_file = os.path.join(base_path, "antique_docs_clean.csv")  # ✅ تم تعديله حسب طلبك

# === تحميل البيانات ===
print("📦 تحميل النموذج والمصفوفة...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"✅ شكل المصفوفة: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

# === تقليل الأبعاد باستخدام UMAP ===
print("📉 تقليل الأبعاد باستخدام UMAP...")
umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]
reduced_data = umap.fit_transform(sample_matrix)

# === عنقدة باستخدام KMeans ===
print("🔄 تنفيذ MiniBatchKMeans...")
n_clusters = 5
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
labels = kmeans.fit_predict(reduced_data)

# === عرض الكلمات المفتاحية لكل عنقود ===
print("\n📌 أهم الكلمات لكل عنقود:")
terms = vectorizer.get_feature_names_out()
for i in range(n_clusters):
    print(f"\nCluster {i}:")
    cluster_indices = np.where(labels == i)[0]
    cluster_matrix = sample_matrix[cluster_indices].mean(axis=0).A1
    top_indices = cluster_matrix.argsort()[::-1][:10]
    top_terms = [terms[j] for j in top_indices]
    print(", ".join(top_terms))

# === تقييم العنقدة ===
print("\n📈 تقييم العنقدة:")
print(f"Silhouette Score: {silhouette_score(reduced_data, labels):.3f}")
print(f"Davies-Bouldin Index: {davies_bouldin_score(reduced_data, labels):.3f}")
print(f"Calinski-Harabasz Index: {calinski_harabasz_score(reduced_data, labels):.3f}")

# === رسم العنقود ===
print("📊 رسم العنقود...")
plt.figure(figsize=(10, 7))
colors = plt.cm.get_cmap('tab10', n_clusters)
for i in range(n_clusters):
    cluster_points = reduced_data[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                label=f"Cluster {i}", s=30, alpha=0.6, color=colors(i))

plt.title("🔵 توزيع الوثائق حسب العنقود")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend()
plt.grid(True)
plt.tight_layout()

plot_path = os.path.join(base_path, "clusters_plot.png")
plt.savefig(plot_path)
plt.show()

# === حفظ نتائج التجميع كـ CSV ===
print("\n💾 حفظ نتائج التجميع...")
docs_df = pd.DataFrame({"doc_id": np.array(doc_ids)[sample_indices]})
docs_df["cluster_label"] = labels
results_path = os.path.join(base_path, "tfidf_clusters.csv")
docs_df.to_csv(results_path, index=False)
print(f"📁 تم حفظ نتائج التجميع في: {results_path}")

# === حفظ نموذج UMAP ===
umap_model_path = os.path.join(base_path, "umap_model.joblib")
joblib.dump(umap, umap_model_path)
print(f"💾 تم حفظ نموذج UMAP في: {umap_model_path}")

# === حفظ نموذج الكلاستر KMeans ===
cluster_model_path = os.path.join(base_path, "cluster_model.joblib")
joblib.dump(kmeans, cluster_model_path)
print(f"💾 تم حفظ نموذج KMeans في: {cluster_model_path}")

print(f"\n✅ التحليل اكتمل. تم حفظ المخطط في '{plot_path}'")
```
