```python
import re
import os
import joblib
import numpy as np
import pandas as pd
from scipy.sparse import save_npz
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
import nltk

# تحميل موارد NLTK
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")
```

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()

def custom_preprocessor(text):
    if text is None:
        return ""
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text

def custom_tokenizer(text):
    text = custom_preprocessor(text)
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    tokens = [stemmer.stem(t) for t in tokens]
    return tokens
```

```python
# اختر "antique" أو "trec_tot"
dataset = "antique"

base_path = os.path.join("data", dataset)
os.makedirs(base_path, exist_ok=True)

doc_file = os.path.join(base_path, "antique_docs_clean.csv") if dataset == "antique" else os.path.join(base_path, "tot_docs_clean.csv")
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")
```

```python
df = pd.read_csv(doc_file)
texts = df["doc_text"].astype(str).tolist()

vectorizer = TfidfVectorizer(
    preprocessor=custom_preprocessor,
    tokenizer=custom_tokenizer,
    lowercase=False,
    max_df=0.9,
    min_df=5
)

tfidf_matrix = vectorizer.fit_transform(texts)

joblib.dump(vectorizer, vectorizer_path)
save_npz(matrix_path, tfidf_matrix)

print("✅ تم حفظ vectorizer و tfidf_matrix بنجاح")
```

```python
# tfidf_clustering_analysis.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

import umap
import joblib
from scipy.sparse import load_npz

# ---------- إعداد عام ----------
dataset = "antique"  # ← غيّر إلى "trec_tot" إن أردت التحليل على trec_tot

vectorizer_path = f"data/{dataset}/tfidf_vectorizer.pkl"
matrix_path = f"data/{dataset}/tfidf_docs_matrix.npz"
docs_path = f"data/{dataset}/{dataset}_docs_clean.csv"

# ---------- تحميل البيانات ----------
print("🔄 تحميل الملفات...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(matrix_path)
docs_df = pd.read_csv(docs_path)

# التأكد من أن doc_id موجودة
if "doc_id" not in docs_df.columns:
    docs_df["doc_id"] = docs_df.index.astype(str)

# ---------- أخذ عينة (اختياري) ----------
max_docs = 5000
if tfidf_matrix.shape[0] > max_docs:
    print(f"📉 تقليل العينة إلى {max_docs} وثيقة...")
    tfidf_matrix = tfidf_matrix[:max_docs]
    docs_df = docs_df.iloc[:max_docs]

# ---------- تقليل الأبعاد باستخدام UMAP ----------
print("📊 تطبيق UMAP لتقليل الأبعاد...")
umap_model = umap.UMAP(n_neighbors=15, min_dist=0.1, metric='cosine', random_state=42)
umap_embeddings = umap_model.fit_transform(tfidf_matrix)

# ---------- التجميع باستخدام KMeans ----------
n_clusters = 5
print(f"🔗 تنفيذ KMeans بـ {n_clusters} عناقيد...")
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(tfidf_matrix)

# ---------- تقييم التجميع ----------
print("📈 تقييم العناقيد:")
silhouette = silhouette_score(tfidf_matrix, cluster_labels)
calinski = calinski_harabasz_score(tfidf_matrix.toarray(), cluster_labels)
davies = davies_bouldin_score(tfidf_matrix.toarray(), cluster_labels)

print(f"🟢 Silhouette Score: {silhouette:.4f}")
print(f"🟢 Calinski-Harabasz Score: {calinski:.2f}")
print(f"🟢 Davies-Bouldin Score: {davies:.4f} (الأقل أفضل)")

# ---------- تحليل الكلمات المفتاحية في كل Cluster ----------
print("🧠 تحليل أهم الكلمات في كل عنقود:")
terms = np.array(vectorizer.get_feature_names_out())
cluster_centers = kmeans.cluster_centers_

for i in range(n_clusters):
    print(f"\n🔹 Cluster {i}:")
    top_indices = cluster_centers[i].argsort()[::-1][:10]
    top_terms = terms[top_indices]
    print(", ".join(top_terms))

# ---------- عرض النتائج ----------
print("📍 رسم توزيع الوثائق حسب العناقيد...")
plt.figure(figsize=(10, 7))
palette = sns.color_palette("hsv", n_clusters)

for i in range(n_clusters):
    indices = cluster_labels == i
    plt.scatter(
        umap_embeddings[indices, 0],
        umap_embeddings[indices, 1],
        label=f"Cluster {i}",
        s=20,
        alpha=0.6,
        color=palette[i]
    )

plt.title(f"TF-IDF + UMAP + KMeans Clustering ({dataset})", fontsize=14)
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === الإعدادات الأساسية ===
dataset = "trec_tot"  # أو antique
project_root = os.getcwd()
base_path = os.path.join(project_root, 'data', dataset)
vectorizer_path = os.path.join(base_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_path, "tfidf_docs_matrix.npz")
doc_file = os.path.join(base_path, "tot_docs_clean.csv")  # ✅ تم تعديله حسب طلبك

# === تحميل البيانات ===
print("📦 تحميل النموذج والمصفوفة...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"✅ شكل المصفوفة: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

# === تقليل الأبعاد باستخدام UMAP ===
print("📉 تقليل الأبعاد باستخدام UMAP...")
umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]
reduced_data = umap.fit_transform(sample_matrix)

# === عنقدة باستخدام KMeans ===
print("🔄 تنفيذ MiniBatchKMeans...")
n_clusters = 5
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
labels = kmeans.fit_predict(reduced_data)

# === عرض الكلمات المفتاحية لكل عنقود ===
print("\n📌 أهم الكلمات لكل عنقود:")
terms = vectorizer.get_feature_names_out()
for i in range(n_clusters):
    print(f"\nCluster {i}:")
    cluster_indices = np.where(labels == i)[0]
    cluster_matrix = sample_matrix[cluster_indices].mean(axis=0).A1
    top_indices = cluster_matrix.argsort()[::-1][:10]
    top_terms = [terms[j] for j in top_indices]
    print(", ".join(top_terms))

# === تقييم العنقدة ===
print("\n📈 تقييم العنقدة:")
print(f"Silhouette Score: {silhouette_score(reduced_data, labels):.3f}")
print(f"Davies-Bouldin Index: {davies_bouldin_score(reduced_data, labels):.3f}")
print(f"Calinski-Harabasz Index: {calinski_harabasz_score(reduced_data, labels):.3f}")

# === رسم العنقود ===
print("📊 رسم العنقود...")
plt.figure(figsize=(10, 7))
colors = plt.cm.get_cmap('tab10', n_clusters)
for i in range(n_clusters):
    cluster_points = reduced_data[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                label=f"Cluster {i}", s=30, alpha=0.6, color=colors(i))

plt.title("🔵 توزيع الوثائق حسب العنقود")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend()
plt.grid(True)
plt.tight_layout()

plot_path = os.path.join(base_path, "clusters_plot.png")
plt.savefig(plot_path)
plt.show()

# === حفظ نتائج التجميع كـ CSV ===
print("\n💾 حفظ نتائج التجميع...")
docs_df = pd.DataFrame({"doc_id": np.array(doc_ids)[sample_indices]})
docs_df["cluster_label"] = labels
results_path = os.path.join(base_path, "tfidf_clusters.csv")
docs_df.to_csv(results_path, index=False)
print(f"📁 تم حفظ نتائج التجميع في: {results_path}")

# === حفظ نموذج UMAP ===
umap_model_path = os.path.join(base_path, "umap_model.joblib")
joblib.dump(umap, umap_model_path)
print(f"💾 تم حفظ نموذج UMAP في: {umap_model_path}")

# === حفظ نموذج الكلاستر KMeans ===
cluster_model_path = os.path.join(base_path, "cluster_model.joblib")
joblib.dump(kmeans, cluster_model_path)
print(f"💾 تم حفظ نموذج KMeans في: {cluster_model_path}")

print(f"\n✅ التحليل اكتمل. تم حفظ المخطط في '{plot_path}'")
```

```python
import pandas as pd
import requests
import os

dataset = "trec_tot"
top_k = 10

queries_file = f"data/{dataset}/{dataset}_queries.csv"
df_queries = pd.read_csv(queries_file)

run_before_file = f"runs/{dataset}_before_refinement.csv"
run_after_file = f"runs/{dataset}_after_refinement.csv"

if not os.path.exists("runs"):
    os.makedirs("runs")
```

```python
run_before = []

for _, row in df_queries.iterrows():
    qid = str(row["query_id"])
    query_text = row["query_text"]  # تأكد من اسم العمود لديك
    
    # استدعاء خدمة البحث بدون تحسين
    response = requests.post(
        "http://127.0.0.1:8001/query-match",
        json={"query": query_text, "top_k": top_k}
    )
    data = response.json()
    
    if "results" in data:
        for doc in data["results"]:
            run_before.append({
                "query_id": qid,
                "doc_id": doc["doc_id"],
                "score": doc["score"]
            })
    else:
        print(f"Error in response for query {qid}: {data}")

df_run_before = pd.DataFrame(run_before)
df_run_before.to_csv(run_before_file, index=False)
print(f"✅ تم حفظ النتائج بدون تحسين في: {run_before_file}")
```

```python
run_after = []

for _, row in df_queries.iterrows():
    qid = str(row["query_id"])
    query_text = row["query_text"]
    
    # تحسين الاستعلام أولاً
    response_refine = requests.post(
        f"http://127.0.0.1:8000/refine-query?dataset={dataset}",
        json={"query": query_text, "options": {"spelling_correction": True, "query_expansion": True, "query_suggestion": True}}
    )
    refined_query = response_refine.json().get("refined_query", query_text)
    
    # بحث على الاستعلام المحسن
    response_search = requests.post(
        "http://127.0.0.1:8001/query-match",
        json={"query": refined_query, "top_k": top_k}
    )
    data = response_search.json()
    
    if "results" in data:
        for doc in data["results"]:
            run_after.append({
                "query_id": qid,
                "doc_id": doc["doc_id"],
                "score": doc["score"]
            })
    else:
        print(f"Error in response for query {qid}: {data}")

df_run_after = pd.DataFrame(run_after)
df_run_after.to_csv(run_after_file, index=False)
print(f"✅ تم حفظ النتائج بعد تحسين الاستعلام في: {run_after_file}")
```

```python
import pandas as pd

def convert_csv_to_trec(input_csv_path, output_trec_path, method_name="tfidf"):
    df = pd.read_csv(input_csv_path)
    with open(output_trec_path, "w") as f:
        for _, row in df.iterrows():
            line = f"{row['query_id']} Q0 {row['doc_id']} {row['score']} {method_name}\n"
            f.write(line)
```

```python
convert_csv_to_trec("runs/trec_tot_before_refinement.csv", "runs/trec_tot_before_refinement.trec", method_name="before")
convert_csv_to_trec("runs/trec_tot_after_refinement.csv", "runs/trec_tot_after_refinement.trec", method_name="after")
```

```python
import pandas as pd
from ranx import Run

def load_trec_file_as_run(path, method_name="tfidf"):
    df = pd.read_csv(path, sep=" ", names=["query_id", "Q0", "doc_id", "score", "method"])
    run = Run.from_dict({
        str(qid): {str(doc_id): float(score) for _, doc_id, score in group[["doc_id", "score"]].itertuples()}
        for qid, group in df.groupby("query_id")
    }, name=method_name)
    return run
```

```python
run_before = load_trec_file_as_run("runs/trec_tot_before_refinement.trec", method_name="before")
run_after = load_trec_file_as_run("runs/trec_tot_after_refinement.trec", method_name="after")
```

```python
import pandas as pd
from ranx import Qrels, Run, evaluate

# تحميل ملفات qrels و run
qrels_df = pd.read_csv("data/trec_tot/trec_tot_qrels.csv")
run_before_df = pd.read_csv("runs/trec_tot_before_refinement.csv")
run_after_df = pd.read_csv("runs/trec_tot_after_refinement.csv")

# بناء كائنات Qrels و Run من بيانات DataFrame
def build_ranx_dict(df):
    ranx_dict = {}
    for _, row in df.iterrows():
        qid = str(row["query_id"])
        docid = str(row["doc_id"])
        score = float(row.get("score", 1))  # بعض ملفات run قد لا تحوي score
        if qid not in ranx_dict:
            ranx_dict[qid] = {}
        ranx_dict[qid][docid] = score
    return ranx_dict

qrels = build_ranx_dict(qrels_df)
run_before = build_ranx_dict(run_before_df)
run_after = build_ranx_dict(run_after_df)

# تحويل إلى كائنات ranx
qrels_obj = Qrels(qrels)
run_before_obj = Run(run_before)
run_after_obj = Run(run_after)

# تنفيذ التقييم مع MAP كمثال
metrics = ["map", "ndcg", "recip_rank"]

results_before = evaluate(qrels_obj, run_before_obj, metrics)
results_after = evaluate(qrels_obj, run_after_obj, metrics)

print("Evaluation before refinement:")
print(results_before)

print("\nEvaluation after refinement:")
print(results_after)
```

```python
from ranx import Qrels, Run, evaluate
import pandas as pd

# تحميل ملفات qrels و run قبل وبعد التحسين
df_qrels = pd.read_csv("data/trec_tot/trec_tot_qrels.csv")
df_run_before = pd.read_csv("runs/trec_tot_before_refinement.csv")
df_run_after = pd.read_csv("runs/trec_tot_after_refinement.csv")

# تحويل qrels إلى dict بصيغة {query_id: {doc_id: relevance}}
qrels_dict = {}
for _, row in df_qrels.iterrows():
    qid = str(row["query_id"])
    docid = str(row["doc_id"])
    rel = int(row["relevance"])
    if qid not in qrels_dict:
        qrels_dict[qid] = {}
    qrels_dict[qid][docid] = rel

# تحويل run قبل التحسين إلى dict بصيغة {query_id: {doc_id: score}}
run_before_dict = {}
for _, row in df_run_before.iterrows():
    qid = str(row["query_id"])
    docid = str(row["doc_id"])
    score = float(row["score"])
    if qid not in run_before_dict:
        run_before_dict[qid] = {}
    run_before_dict[qid][docid] = score

# تحويل run بعد التحسين إلى dict بصيغة {query_id: {doc_id: score}}
run_after_dict = {}
for _, row in df_run_after.iterrows():
    qid = str(row["query_id"])
    docid = str(row["doc_id"])
    score = float(row["score"])
    if qid not in run_after_dict:
        run_after_dict[qid] = {}
    run_after_dict[qid][docid] = score

# إنشاء كائنات Qrels و Run
qrels_obj = Qrels(qrels_dict)
run_before_obj = Run(run_before_dict)
run_after_obj = Run(run_after_dict)

# تعيين المقاييس التي تريد تقييمها
metrics = ["map", "ndcg", "mrr"]

# تنفيذ التقييم مع make_comparable=True
results_before = evaluate(qrels_obj, run_before_obj, metrics, make_comparable=True)
results_after = evaluate(qrels_obj, run_after_obj, metrics, make_comparable=True)

print("Evaluation before refinement:")
print(results_before)

print("\nEvaluation after refinement:")
print(results_after)
```

```python
import os
import json
import pandas as pd
import numpy as np
from scipy import sparse
import joblib

# ==== إعداد المسارات ====
dataset = "trec_tot"  # غيّر حسب الحاجة "antique"
base_dir = os.path.abspath(".")

vectorizer_path = os.path.join(base_dir, 'offline_indexing_service', 'data', dataset, 'tfidf_vectorizer.pkl')
matrix_path = os.path.join(base_dir, 'offline_indexing_service', 'data', dataset, 'tfidf_docs_matrix.npz')
doc_ids_path = os.path.join(base_dir, 'data', dataset, 'tfidf_doc_ids.json')
clusters_csv_path = os.path.join(base_dir, 'offline_indexing_service', 'data', dataset, 'tfidf_clusters.csv')

# ==== تحميل البيانات ====
print("تحميل الملفات...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = sparse.load_npz(matrix_path)
with open(doc_ids_path, "r") as f:
    doc_ids = json.load(f)
df_clusters = pd.read_csv(clusters_csv_path)

print(f"عدد doc_ids: {len(doc_ids)}")
print(f"عدد صفوف الكلستر: {len(df_clusters)}")

# ==== تحقق 1: هل جميع doc_ids في الكلستر موجودين في قائمة doc_ids؟ ====
missing_in_doc_ids = set(df_clusters['doc_id']) - set(doc_ids)
print(f"عدد doc_ids في الكلستر غير موجودة في doc_ids: {len(missing_in_doc_ids)}")
if missing_in_doc_ids:
    print("أمثلة عليها:", list(missing_in_doc_ids)[:10])

# ==== تحقق 2: هل doc_ids فريدة؟ ====
if len(doc_ids) != len(set(doc_ids)):
    print("تحذير: توجد doc_ids مكررة في ملف tfidf_doc_ids.json")
else:
    print("doc_ids فريدة")

# ==== بناء قاموس doc_id -> index في المصفوفة ====
docid_to_index = {doc_id: idx for idx, doc_id in enumerate(doc_ids)}

# ==== تحقق 3: هل جميع doc_ids في الكلستر موجودة في القاموس؟ ====
missing_in_index = [doc_id for doc_id in df_clusters['doc_id'] if doc_id not in docid_to_index]
print(f"عدد doc_ids في الكلستر غير موجودة في قاموس docid_to_index: {len(missing_in_index)}")
if missing_in_index:
    print("أمثلة عليها:", missing_in_index[:10])

# ==== اختبار تمثيل استعلام افتراضي والدرجات ====
test_query = "test query example"
cleaned_query = test_query  # افتراضياً، استبدل بدالة التنظيف لديك
query_vec = vectorizer.transform([cleaned_query])
scores = (query_vec * tfidf_matrix.T).toarray()[0]  # cosine similarity تقريبي

# ==== تحقق 4: هل توجد NaN أو Inf في scores؟ ====
print("هل يوجد NaN في درجات التشابه؟", np.any(np.isnan(scores)))
print("هل يوجد Inf في درجات التشابه؟", np.any(np.isinf(scores)))

# ==== تحقق 5: هل الطول متساوي؟ ====
print(f"طول scores: {len(scores)} مقابل عدد doc_ids: {len(doc_ids)}")

# ==== تحقق 6: تجميع الدرجات حسب الكلستر ====
doc_cluster_map = dict(zip(df_clusters['doc_id'], df_clusters['cluster_label']))

clusters_scores = {}
for idx, score in enumerate(scores):
    doc_id = doc_ids[idx]
    cluster_label = doc_cluster_map.get(doc_id, -1)
    clusters_scores.setdefault(cluster_label, 0)
    clusters_scores[cluster_label] += score

print(f"مجموع الدرجات حسب الكلستر: {clusters_scores}")

# ==== تحقق 7: جمع النتائج لكل كلستر مع التحقق من وجود doc_id في القاموس ====
sorted_clusters = sorted(clusters_scores.items(), key=lambda x: x[1], reverse=True)

results = []
for cluster_label, _ in sorted_clusters:
    cluster_docs = [doc_id for doc_id, cl_label in doc_cluster_map.items() if cl_label == cluster_label]
    cluster_docs_scores = []
    for doc_id in cluster_docs:
        idx = docid_to_index.get(doc_id, None)
        if idx is not None:
            score = scores[idx]
            if np.isnan(score) or np.isinf(score):
                print(f"score غير صالح للوثيقة {doc_id}: {score}")
                continue
            cluster_docs_scores.append((doc_id, score))
        else:
            print(f"doc_id {doc_id} غير موجود في قاموس docid_to_index")
    cluster_docs_scores.sort(key=lambda x: x[1], reverse=True)
    results.extend(cluster_docs_scores)

print(f"عدد النتائج بعد التجميع: {len(results)}")
print("أعلى 10 نتائج:", results[:10])
```
