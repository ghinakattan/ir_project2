```python
import requests
import json
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, HTML

# Ø¥Ø¹Ø¯Ø§Ø¯Ø§Øª Ø¹Ø§Ù…Ø©
BASE_URL = "http://localhost:8002/evaluate"
DATASET = "trec_tot"
QRELS_FILE = "trec_tot_qrels.csv"
TOP_K = 10

representations = [
    {"name": "TF-IDF", "run_file": "tfidf_run_top10.csv", "results_file": "tfidf_eval.json"},
    {"name": "Embedding", "run_file": "embedding_run_top10.csv", "results_file": "embedding_eval.json"},
    {"name": "Hybrid", "run_file": "hybrid_run_top10.csv", "results_file": "hybrid_eval.json"},
]

all_results = {}

for rep in representations:
    print(f"\nðŸ“Œ Ø§Ù„ØªÙ‚ÙŠÙŠÙ… Ù„ØªÙ…Ø«ÙŠÙ„: {rep['name']}")
    response = requests.get(BASE_URL, params={
        "dataset": DATASET,
        "qrels_file": QRELS_FILE,
        "run_file": rep["run_file"],
        "top_k": TOP_K,
        "results_file": rep["results_file"]
    })

    if response.status_code == 200:
        result = response.json()
        metrics = result["results"]
        all_results[rep["name"]] = metrics
        for metric, value in metrics.items():
            print(f"âœ… {metric}: {round(value, 4)}")
    else:
        print(f"âŒ ÙØ´Ù„ Ø§Ù„ØªÙ‚ÙŠÙŠÙ…: {response.status_code} - {response.text}")

df_results = pd.DataFrame(all_results).T
metric_order = ['map', 'mrr', 'ndcg@10', 'precision@10', 'recall@10']
cols = [col for col in metric_order if col in df_results.columns]
df_results = df_results[cols]
df_results = df_results.round(4)

# Ø¹Ø±Ø¶ Ø§Ù„Ø¬Ø¯ÙˆÙ„ Ù…Ø¹ RTL Ø¨Ø¯ÙˆÙ† .render()
display(df_results.style.set_table_attributes('style="direction: rtl; text-align: right;"'))

plt.figure(figsize=(12, 6))
ax = df_results.plot(kind='bar', colormap='viridis')
plt.title("ðŸ“Š Ù…Ù‚Ø§Ø±Ù†Ø© Ø§Ù„Ø£Ø¯Ø§Ø¡ Ø¨ÙŠÙ† Ø§Ù„ØªÙ…Ø«ÙŠÙ„Ø§Øª Ø§Ù„Ù…Ø®ØªÙ„ÙØ©", fontsize=16, loc='right')
plt.ylabel("Ø§Ù„Ù‚ÙŠÙ…Ø©", fontsize=12)
plt.xlabel("Ø§Ù„ØªÙ…Ø«ÙŠÙ„", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Ø§Ù„Ù…Ù‚ÙŠØ§Ø³", loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

best_map_method = df_results['map'].idxmax()

summary_html = f"""
<div style="direction: rtl; text-align: right; font-size:16px; line-height: 1.6;">
ØªØ´ÙŠØ± Ù†ØªØ§Ø¦Ø¬ Ø§Ù„ØªÙ‚ÙŠÙŠÙ… Ø¥Ù„Ù‰ Ø£Ù† ØªÙ…Ø«ÙŠÙ„ <b>{best_map_method}</b> Ø­Ù‚Ù‚ Ø£ÙØ¶Ù„ Ø£Ø¯Ø§Ø¡ Ù…Ù† Ø­ÙŠØ« Ù…Ù‚ÙŠØ§Ø³ MAPØŒ ÙˆÙ‡Ùˆ Ø§Ù„Ù…Ù‚ÙŠØ§Ø³ Ø§Ù„Ø£Ø³Ø§Ø³ÙŠ ÙÙŠ Ù‡Ø°Ø§ Ø§Ù„Ù…Ø´Ø±ÙˆØ¹.
ÙƒÙ…Ø§ Ø£Ø¸Ù‡Ø±Øª Ø§Ù„Ù…Ù‚Ø§ÙŠÙŠØ³ Ø§Ù„Ø£Ø®Ø±Ù‰ Ù…Ø«Ù„ MRRØªÙÙˆÙ‚Ù‹Ø§ ÙˆØ§Ø¶Ø­Ù‹Ø§ Ø£ÙŠØ¶Ù‹Ø§ Ù„Ù†ÙØ³ Ø§Ù„ØªÙ…Ø«ÙŠÙ„ ÙÙŠ Ø¨Ø¹Ø¶ Ø§Ù„Ø­Ø§Ù„Ø§ØªØŒ Ù…Ù…Ø§ ÙŠØ¹Ø²Ø² Ù…ÙˆØ«ÙˆÙ‚ÙŠØªÙ‡.

<ul>
<li>ØªÙ…Ø«ÙŠÙ„ TF-IDF Ø£Ø¸Ù‡Ø± Ø£Ø¯Ø§Ø¡Ù‹ Ø¬ÙŠØ¯Ù‹Ø§ ÙÙŠ Ø§Ù„Ù…Ù‚Ø§ÙŠÙŠØ³ Ø§Ù„ØªÙ‚Ù„ÙŠØ¯ÙŠØ© Ø§Ù„Ù…Ø±ØªØ¨Ø·Ø© Ø¨Ø§Ù„Ø¯Ù‚Ø©.</li>
<li>Ø¨ÙŠÙ†Ù…Ø§ Ø£Ø¸Ù‡Ø± ØªÙ…Ø«ÙŠÙ„ Embedding Ø£Ø¯Ø§Ø¡Ù‹ Ù…ØªÙˆØ§Ø²Ù†Ù‹Ø§ Ù„ÙƒÙ†Ù‡ Ù„Ù… ÙŠØªÙÙˆÙ‚ Ø¹Ù„Ù‰ Ø§Ù„Ù‡Ø¬ÙŠÙ†.</li>
<li>Ø§Ù„ØªÙ…Ø«ÙŠÙ„ Ø§Ù„Ù‡Ø¬ÙŠÙ† Hybrid Ù†Ø¬Ø­ ÙÙŠ Ø§Ù„Ø§Ø³ØªÙØ§Ø¯Ø© Ù…Ù† Ù†Ù‚Ø§Ø· Ø§Ù„Ù‚ÙˆØ© ÙÙŠ ÙƒÙ„Ø§ Ø§Ù„ØªÙ…Ø«ÙŠÙ„ÙŠÙ†ØŒ Ù…Ø­Ù‚Ù‚Ù‹Ø§ Ù†ØªØ§Ø¦Ø¬ Ù‚ÙˆÙŠØ© ÙˆÙ…ØªÙÙˆÙ‚Ø© ÙÙŠ Ù…Ø¹Ø¸Ù… Ø§Ù„Ù…Ù‚Ø§ÙŠÙŠØ³.</li>
</ul>

Ø¨Ù†Ø§Ø¡Ù‹ Ø¹Ù„ÙŠÙ‡ØŒ ÙŠÙ…ÙƒÙ† Ø§Ø¹ØªØ¨Ø§Ø± Ø§Ù„ØªÙ…Ø«ÙŠÙ„ Ø§Ù„Ù‡Ø¬ÙŠÙ† Ø§Ø®ØªÙŠØ§Ø±Ù‹Ø§ Ù…ÙˆØµÙ‰ Ø¨Ù‡ Ù„Ù„ØªØ·Ø¨ÙŠÙ‚Ø§Øª Ø§Ù„ÙØ¹Ù„ÙŠØ©ØŒ Ø®ØµÙˆØµÙ‹Ø§ Ø¹Ù†Ø¯Ù…Ø§ ØªÙƒÙˆÙ† Ø§Ù„Ø¬ÙˆØ¯Ø© ÙÙŠ Ø§Ø³ØªØ±Ø¬Ø§Ø¹ Ø§Ù„Ù†ØªØ§Ø¦Ø¬ Ø£ÙˆÙ„ÙˆÙŠØ©.
</div>
"""

display(HTML(summary_html))
```
