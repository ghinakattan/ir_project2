```python
import os
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import (
    silhouette_score,
    davies_bouldin_score,
    calinski_harabasz_score
)
from umap import UMAP
from scipy.sparse import load_npz

# === ÿßŸÑÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑÿ£ÿ≥ÿßÿ≥Ÿäÿ© ===
dataset = "antique"  # ÿ£Ÿà "antique"
project_root = os.getcwd()

# ÿ≠ÿ≥ÿ® ÿßŸÑŸÖÿ≥ÿßÿ±ÿßÿ™ ÿßŸÑÿ™Ÿä ÿ£ÿ±ÿ≥ŸÑÿ™Ÿáÿßÿå ŸÖŸÑŸÅÿßÿ™ tfidf ÿ∂ŸÖŸÜ offline_indexing_service
base_offline_indexing_path = os.path.join(project_root, "offline_indexing_service", "data", dataset)

vectorizer_path = os.path.join(base_offline_indexing_path, "tfidf_vectorizer.pkl")
tfidf_matrix_path = os.path.join(base_offline_indexing_path, "tfidf_docs_matrix.npz")

# ŸÖŸÑŸÅÿßÿ™ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ÿßŸÑÿ£ÿµŸÑŸäÿ© ŸàÿßŸÑŸÖŸÜÿ∏ŸÅÿ© ÿ∂ŸÖŸÜ ŸÖÿ¨ŸÑÿØ data ŸÖÿ®ÿßÿ¥ÿ±ÿ© (ŸàŸÑŸäÿ≥ ÿ∂ŸÖŸÜ offline_indexing_service)
base_data_path = os.path.join(project_root, "data", dataset)

if dataset == "trec_tot":
    doc_file = os.path.join(base_data_path, "trec_tot_docs_clean.csv")
elif dataset == "antique":
    doc_file = os.path.join(base_data_path, "antique_docs_clean.csv")
else:
    doc_file = None  # ÿ£Ÿà ÿ±ŸÅÿπ ÿßÿ≥ÿ™ÿ´ŸÜÿßÿ° ŸáŸÜÿß

print("vectorizer_path:", vectorizer_path)
print("tfidf_matrix_path:", tfidf_matrix_path)
print("doc_file:", doc_file)

# === ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ===
print("üì¶ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸàÿßŸÑŸÖÿµŸÅŸàŸÅÿ©...")
vectorizer = joblib.load(vectorizer_path)
tfidf_matrix = load_npz(tfidf_matrix_path)

print(f"‚úÖ ÿ¥ŸÉŸÑ ÿßŸÑŸÖÿµŸÅŸàŸÅÿ©: {tfidf_matrix.shape}")
doc_ids = pd.read_csv(doc_file)["doc_id"].astype(str).tolist()

 # === ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP (ŸÑŸÑÿπÿ±ÿ∂ ŸÅŸÇÿ∑) ===
print("üìâ ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ£ÿ®ÿπÿßÿØ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ UMAP...")
sample_size = min(5000, tfidf_matrix.shape[0])
sample_indices = np.random.choice(tfidf_matrix.shape[0], sample_size, replace=False)
sample_matrix = tfidf_matrix[sample_indices]

umap = UMAP(n_components=2, n_neighbors=30, min_dist=0.0, metric='cosine', random_state=42)
reduced_data = umap.fit_transform(sample_matrix)

# === ÿπŸÜŸÇÿØÿ© ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ KMeans ÿπŸÑŸâ ŸÉÿßŸÖŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ (ŸàŸÑŸäÿ≥ ŸÅŸÇÿ∑ ÿßŸÑÿπŸäŸÜÿ©) ===
print("üîÑ ÿ™ŸÜŸÅŸäÿ∞ MiniBatchKMeans ÿπŸÑŸâ ŸÉÿßŸÖŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ...")
n_clusters = 5
kmeans = MiniBatchKMeans(n_clusters=n_clusters, batch_size=512, random_state=42)
full_labels = kmeans.fit_predict(tfidf_matrix)


# === ÿ™ŸÇŸäŸäŸÖ ÿ¨ŸàÿØÿ© ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ===
print("\nüìä ÿ™ŸÇŸäŸäŸÖ ÿ¨ŸàÿØÿ© ÿßŸÑÿπŸÜÿßŸÇŸäÿØ:")
try:
    silhouette = silhouette_score(tfidf_matrix, full_labels)
    davies = davies_bouldin_score(tfidf_matrix.toarray(), full_labels)
    calinski = calinski_harabasz_score(tfidf_matrix.toarray(), full_labels)

    print(f"‚úÖ Silhouette Score: {silhouette:.4f}")
    print(f"‚úÖ Davies-Bouldin Score: {davies:.4f}")
    print(f"‚úÖ Calinski-Harabasz Score: {calinski:.4f}")

    # ÿ≠ŸÅÿ∏ ÿßŸÑÿ™ŸÇŸäŸäŸÖÿßÿ™ ŸÅŸä ŸÖŸÑŸÅ
    eval_path = os.path.join(base_offline_indexing_path, "cluster_evaluation.json")
    import json
    with open(eval_path, "w") as f:
        json.dump({
            "silhouette_score": silhouette,
            "davies_bouldin_score": davies,
            "calinski_harabasz_score": calinski
        }, f, indent=2)

    print(f"üíæ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ŸÇŸäŸäŸÖ ŸÅŸä: {eval_path}")
except Exception as e:
    print(f"‚ö†Ô∏è ÿ≠ÿØÿ´ ÿÆÿ∑ÿ£ ÿ£ÿ´ŸÜÿßÿ° ÿßŸÑÿ™ŸÇŸäŸäŸÖ: {e}")


# === ÿπÿ±ÿ∂ ÿ£ŸáŸÖ ÿßŸÑŸÉŸÑŸÖÿßÿ™ ŸÑŸÉŸÑ ÿπŸÜŸÇŸàÿØ (ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ŸÉÿßŸÖŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™) ===
print("\nüìå ÿ£ŸáŸÖ ÿßŸÑŸÉŸÑŸÖÿßÿ™ ŸÑŸÉŸÑ ÿπŸÜŸÇŸàÿØ:")
terms = vectorizer.get_feature_names_out()
for i in range(n_clusters):
    cluster_indices = np.where(full_labels == i)[0]
    cluster_matrix = tfidf_matrix[cluster_indices].mean(axis=0).A1
    top_indices = cluster_matrix.argsort()[::-1][:10]
    top_terms = [terms[j] for j in top_indices]
    print(f"\nCluster {i}:")
    print(", ".join(top_terms))

# === ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± ŸÑŸÉŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ===
print("\nüíæ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ...")
docs_df = pd.DataFrame({
    "doc_id": doc_ids,  # ÿ¨ŸÖŸäÿπ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ
    "cluster_label": full_labels
})
results_path = os.path.join(base_offline_indexing_path, "tfidf_clusters.csv")
docs_df.to_csv(results_path, index=False)
print(f"üìÅ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ŸÅŸä: {results_path}")

# === ÿ≠ŸÅÿ∏ ÿßŸÑŸÖÿÆÿ∑ÿ∑ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿßŸÑÿπŸäŸÜÿ© ŸÅŸÇÿ∑ (ŸÉŸÖÿß ŸáŸà) ===
print("üìä ÿ±ÿ≥ŸÖ ÿßŸÑÿπŸÜŸÇŸàÿØ...")
plt.figure(figsize=(10, 7))
colors = plt.cm.get_cmap('tab10', n_clusters)
sample_labels = full_labels[sample_indices]  # ÿÆÿ∞ ÿßŸÑÿ™ÿµŸÜŸäŸÅÿßÿ™ ŸÖŸÜ ŸÉÿßŸÖŸÑ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ

for i in range(n_clusters):
    cluster_points = reduced_data[sample_labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                label=f"Cluster {i}", s=30, alpha=0.6, color=colors(i))

plt.title("üîµ ÿ™Ÿàÿ≤Ÿäÿπ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ ÿ≠ÿ≥ÿ® ÿßŸÑÿπŸÜŸÇŸàÿØ")
plt.xlabel("UMAP-1")
plt.ylabel("UMAP-2")
plt.legend()
plt.grid(True)
plt.tight_layout()

plot_path = os.path.join(base_offline_indexing_path, "clusters_plot.png")
plt.savefig(plot_path)
plt.show()

# === ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ŸÉŸÄ CSV ===
print("\nüíæ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ...")
docs_df = pd.DataFrame({
    "doc_id": doc_ids,                 # ÿ¨ŸÖŸäÿπ ÿßŸÑŸàÿ´ÿßÿ¶ŸÇ
    "cluster_label": full_labels      # ÿßŸÑÿ™ÿµŸÜŸäŸÅÿßÿ™ ÿßŸÑÿ¨ÿØŸäÿØÿ© ŸÑŸÉŸÑ Ÿàÿ´ŸäŸÇÿ©
})
results_path = os.path.join(base_offline_indexing_path, "tfidf_clusters.csv")
docs_df.to_csv(results_path, index=False)
print(f"üìÅ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜÿ™ÿßÿ¶ÿ¨ ÿßŸÑÿ™ÿ¨ŸÖŸäÿπ ŸÅŸä: {results_path}")

# === ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ UMAP ===
umap_model_path = os.path.join(base_offline_indexing_path, "umap_model.joblib")
joblib.dump(umap, umap_model_path)
print(f"üíæ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ UMAP ŸÅŸä: {umap_model_path}")

# === ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ ÿßŸÑŸÉŸÑÿßÿ≥ÿ™ÿ± KMeans ===
cluster_model_path = os.path.join(base_offline_indexing_path, "cluster_model.joblib")
joblib.dump(kmeans, cluster_model_path)
print(f"üíæ ÿ™ŸÖ ÿ≠ŸÅÿ∏ ŸÜŸÖŸàÿ∞ÿ¨ KMeans ŸÅŸä: {cluster_model_path}")

print(f"\n‚úÖ ÿßŸÑÿ™ÿ≠ŸÑŸäŸÑ ÿßŸÉÿ™ŸÖŸÑ. ÿ™ŸÖ ÿ≠ŸÅÿ∏ ÿßŸÑŸÖÿÆÿ∑ÿ∑ ŸÅŸä '{plot_path}'")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import joblib


import re

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer


from sentence_transformers import SentenceTransformer

# ================================
# ÿ•ÿπÿØÿßÿØÿßÿ™ ÿßŸÑŸÖÿπÿßŸÑÿ¨ÿ© ÿßŸÑŸÜÿµŸäÿ©
# ================================
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()


# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
model_path = 'offline_indexing_service/data/trec_tot/embedding_model/model.joblib'
model = joblib.load(model_path)

# ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ™ŸÖÿ´ŸäŸÑÿßÿ™ Ÿàdoc_ids
embeddings = np.load('offline_indexing_service/data/trec_tot/doc_embeddings.npy')
with open('offline_indexing_service/data/trec_tot/embedding_doc_ids.json', 'r') as f:
    doc_ids = json.load(f)

# ÿ™ŸÖÿ´ŸäŸÑ ÿßŸÑÿßÿ≥ÿ™ÿπŸÑÿßŸÖ (ÿ¨ÿ±ÿ® ÿßÿ≥ÿ™ÿπŸÑÿßŸÖŸÉ)
query = "HOw to buy a book for cooking"

def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    tokens = text.split()
    tokens = [t for t in tokens if t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]
    tokens = [stemmer.stem(t) for t in tokens]
    return " ".join(tokens)

cleaned_query = clean_text(query)
query_vec = model.encode([cleaned_query])  # (1, dim)

# ÿ≠ÿ≥ÿßÿ® ÿßŸÑÿ™ÿ¥ÿßÿ®Ÿá
scores = cosine_similarity(query_vec, embeddings)[0]

print(f"Scores stats:")
print(f" - min: {scores.min()}")
print(f" - max: {scores.max()}")
print(f" - mean: {scores.mean()}")
print(f" - top 10 scores: {sorted(scores, reverse=True)[:10]}")
```
