```python
import requests
import json
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, HTML

# إعدادات عامة
BASE_URL = "http://localhost:8002/evaluate"
DATASET = "trec_tot"
QRELS_FILE = "trec_tot_qrels.csv"
TOP_K = 10

representations = [
    {"name": "TF-IDF", "run_file": "tfidf_run_top10.csv", "results_file": "tfidf_eval.json"},
    {"name": "Embedding", "run_file": "embedding_run_top10.csv", "results_file": "embedding_eval.json"},
    {"name": "Hybrid", "run_file": "hybrid_run_top10.csv", "results_file": "hybrid_eval.json"},
]

all_results = {}

for rep in representations:
    print(f"\n📌 التقييم لتمثيل: {rep['name']}")
    response = requests.get(BASE_URL, params={
        "dataset": DATASET,
        "qrels_file": QRELS_FILE,
        "run_file": rep["run_file"],
        "top_k": TOP_K,
        "results_file": rep["results_file"]
    })

    if response.status_code == 200:
        result = response.json()
        metrics = result["results"]
        all_results[rep["name"]] = metrics
        for metric, value in metrics.items():
            print(f"✅ {metric}: {round(value, 4)}")
    else:
        print(f"❌ فشل التقييم: {response.status_code} - {response.text}")

df_results = pd.DataFrame(all_results).T
metric_order = ['map', 'mrr', 'ndcg@10', 'precision@10', 'recall@10']
cols = [col for col in metric_order if col in df_results.columns]
df_results = df_results[cols]
df_results = df_results.round(4)

# عرض الجدول مع RTL بدون .render()
display(df_results.style.set_table_attributes('style="direction: rtl; text-align: right;"'))

plt.figure(figsize=(12, 6))
ax = df_results.plot(kind='bar', colormap='viridis')
plt.title("📊 مقارنة الأداء بين التمثيلات المختلفة", fontsize=16, loc='right')
plt.ylabel("القيمة", fontsize=12)
plt.xlabel("التمثيل", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="المقياس", loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

best_map_method = df_results['map'].idxmax()
```
